```python
import json
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = []
with open(filepath, "r") as f:
    for line in f:
        try:
            data.append(json.loads(line))
        except json.JSONDecodeError:
            continue

df = pd.DataFrame(data)
df
```

```python
print(f"Total loaded: {len(df)}")
print(df['label'].value_counts())
```

```python
for label in ['benign', 'phishing']:
    sample = df[df['label'] == label].iloc[0]
    print(f"\n{'='*80}")
    print(f"LABEL: {label.upper()}")
    print(f"{'='*80}")
    for col in df.columns:
        val = sample[col]
        if isinstance(val, (dict, list)):
            print(f"\n--- {col} ---")
            print(json.dumps(val, indent=2, default=str)[:2000])  # truncate long fields
        else:
            print(f"\n--- {col} ---")
            print(str(val)[:500])
```

```python
# Find examples with many signals to see the full range
df['signal_count'] = df['signals'].apply(len)
print("Signal count distribution:")
print(df['signal_count'].describe())
print()
print(df.groupby('label')['signal_count'].describe())
```

```python
# Show a benign and phishing example that have the MOST signals
for label in ['benign', 'phishing']:
    subset = df[df['label'] == label]
    idx = subset['signal_count'].idxmax()
    sample = df.loc[idx]
    print(f"\n{'='*80}")
    print(f"LABEL: {label.upper()} | SIGNAL COUNT: {sample['signal_count']}")
    print(f"URL: {sample['requested_url']}")
    print(f"{'='*80}")
    print(f"\n--- signals ---")
    print(json.dumps(sample['signals'], indent=2, default=str))
    print(f"\n--- html (title + forms) ---")
    html = sample['html']
    print(f"Title: {html.get('title')}")
    print(f"Forms: {json.dumps(html.get('forms', []), indent=2, default=str)[:2000]}")
    print(f"\n--- url ---")
    print(json.dumps(sample['url'], indent=2, default=str))
```

```python
# Explode signals into individual rows for analysis
signals_expanded = df[['source_id', 'label', 'signals']].explode('signals')
signals_expanded['signal_id'] = signals_expanded['signals'].apply(lambda x: x['id'] if isinstance(x, dict) else None)
signals_expanded['polarity'] = signals_expanded['signals'].apply(lambda x: x['polarity'] if isinstance(x, dict) else None)
signals_expanded = signals_expanded.dropna(subset=['signal_id'])

# Count of benign vs phishing polarity signals per label
print("Signal polarity counts by site label:")
print(signals_expanded.groupby(['label', 'polarity']).size().unstack(fill_value=0))
print()

# Average per site
print("\nAverage signal counts per site:")
site_counts = signals_expanded.groupby(['source_id', 'label', 'polarity']).size().reset_index(name='count')
print(site_counts.groupby(['label', 'polarity'])['count'].mean().unstack(fill_value=0).round(2))
```

```python
# Most common signals in PHISHING sites (phishing-polarity signals)
phishing_signals = signals_expanded[(signals_expanded['label'] == 'phishing') & (signals_expanded['polarity'] == 'phishing')]
print("TOP PHISHING-POLARITY SIGNALS IN PHISHING SITES:")
print(phishing_signals['signal_id'].value_counts().head(20).to_string())

print(f"\n{'='*80}\n")

# Most common signals in BENIGN sites (benign-polarity signals)
benign_signals = signals_expanded[(signals_expanded['label'] == 'benign') & (signals_expanded['polarity'] == 'benign')]
print("TOP BENIGN-POLARITY SIGNALS IN BENIGN SITES:")
print(benign_signals['signal_id'].value_counts().head(20).to_string())
```

```python
# Also interesting: which phishing-polarity signals appear in BENIGN sites (false positives)?
fp_signals = signals_expanded[(signals_expanded['label'] == 'benign') & (signals_expanded['polarity'] == 'phishing')]
print("PHISHING-POLARITY SIGNALS IN BENIGN SITES (potential false positives):")
print(fp_signals['signal_id'].value_counts().head(20).to_string())

print(f"\n{'='*80}\n")

# And benign-polarity signals in PHISHING sites (signals that fail to catch phishing)
fn_signals = signals_expanded[(signals_expanded['label'] == 'phishing') & (signals_expanded['polarity'] == 'benign')]
print("BENIGN-POLARITY SIGNALS IN PHISHING SITES (missed by these checks):")
print(fn_signals['signal_id'].value_counts().head(20).to_string())
```
